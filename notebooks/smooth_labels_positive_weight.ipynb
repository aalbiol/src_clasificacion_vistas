```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ps=[1,3,5] # positive weight

ys=[0.1]
prob=np.linspace(0.01,0.99,100) # Sigmoidde de los logits
legends=[]
for y in ys:
    for p in ps:
        loss= -y*p*np.log(prob) - (1-y)*np.log(1-prob)
        _=plt.plot(prob,loss)
        plt.xlabel('p')
        plt.ylabel('loss')
        legends.append(f'Cross-entropy loss target={y:.2f} pos={p:.1f}')
plt.grid()
plt.legend(legends)
```
